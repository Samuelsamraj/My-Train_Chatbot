```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import string
import pandas as pd

# Load dataset
# Replace '/content/Train_details_22122017.csv' with the actual filename and path
train_data = pd.read_csv("/content/Train_details_22122017.csv")

# Initialize NLTK components
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

# Preprocess user input
def preprocess_input(user_input):
    # Tokenize input
    tokens = word_tokenize(user_input.lower())
    # Remove punctuation and stop words, and lemmatize tokens
    tokens = [lemmatizer.lemmatize(token) for token in tokens if token not in string.punctuation and token not in stop_words]
    return tokens

# Function to retrieve train information based on user input
def get_train_info(origin, destination):
    # Filter dataset based on origin and destination
    relevant_trains = train_data[(train_data['Station Name'] == origin) & (train_data['Destination Station Name'] == destination)]
    # Extract relevant information (train name, departure time, etc.)
    train_info = relevant_trains[['Train Name', 'Departure Time', 'Arrival time']]
    return train_info

# Main function to interact with the user
def chat():
    print("Welcome to MyTrain Chatbot!")
    while True:
        user_input = input("Please enter your origin and destination stations (e.g., 'Delhi to Mumbai'): ")
        if user_input.lower() == 'exit':
            print("Thank you for using MyTrain Chatbot. Goodbye!")
            break
        else:
            # Preprocess user input
            tokens = preprocess_input(user_input)
            if len(tokens) < 2:
                print("Please provide both origin and destination stations.")
                continue
            else:
                # Assume first two tokens are origin and destination
                origin = tokens[0]
                destination = tokens[1]
                # Retrieve train information
                train_info = get_train_info(origin, destination)
                if train_info.empty:
                    print("Sorry, no trains found for the specified route.")
                else:
                    print("Available trains:")
                    print(train_info)

# Start the chatbot
if __name__ == "__main__":
    chat()
```
